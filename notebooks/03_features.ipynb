```python
import pandas as pd

df = pd.read_csv("../data/raw.csv")

features = [
    "tenure",
    "MonthlyCharges",
    "Contract",
    "InternetService",
    "PaymentMethod",
    "SeniorCitizen",
    "Churn"
]

df = df[features]
df.head()
```

```python
df['Churn'] = df["Churn"].map({"Yes": 1, "No": 0})
```

```python
contract_map = {
    "Month-to-month": 0,
    "One year": 1,
    "Two year": 2
}

df["Contract"] = df["Contract"].map(contract_map)
```

```python
df = pd.get_dummies(df, columns=["InternetService", "PaymentMethod"], drop_first=True)
```

```python
df.head()
df.info()
df.to_csv("../data/cleaned.csv", index=False)
```
